```python

import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
import numpy as np
from PSDE.basis import *
import matplotlib.pyplot as plt
import torch
```

```python

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
partition_size = 100
partition = torch.linspace(0, 5, partition_size, device=device)

st = span_torch_laguerre(partition, 10)
st.shape
```

```python
for x in st:
    plt.plot(partition.cpu().numpy(), x.cpu().numpy())
plt.title('First 10 orthonormal Laguerre functionals on [0,5]')
plt.xlabel('Time')
plt.ylabel('Instant forward rate')
plt.savefig('laguerre.png')
plt.show()
```

```python
st = span_torch(partition, 10)

for x in st:
    plt.plot(partition.cpu().numpy(), x.cpu().numpy())
plt.title('First 10 orthonormal functionals on [0,5]')
plt.xlabel('Time')
plt.ylabel('Instant forward rate')
plt.savefig('exponential.png')
plt.show()
```
